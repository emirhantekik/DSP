```python
%matplotlib notebook
import numpy as np
from scipy.signal import convolve, lfilter
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

n = np.arange(11)
x = np.zeros_like(n)
x[n>=3] = 1
from scipy.signal import convolve, lfilter

## impulse cevabı için lfiltre fonksiyonu kullanılacağından kaynaklı y[n]' katsayıları alınmıştır. 1 ve -1.2
h = [1, -1.2]

# Çıkış işareti (convolve)
y_conv = convolve(x, h, mode='same')

# Çıkış işareti (lfilter)
y_lfilter = lfilter(h, [1], x)

fig = plt.figure()
plt.stem(n, y_conv, use_line_collection=True)
plt.title('Çıkış İşareti (convolve)')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.show()

fig = plt.figure()
plt.stem(n, y_lfilter, use_line_collection=True)
plt.title('Çıkış İşareti (lfilter)')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# giriş işareti x5[n]
n = np.arange(-5, 25)

## Hocaya sordum eğer herhangi bir fonksiyonun kullanılması yasaklanmadıysa kullanabilirsin dedi
## u[n] işareti için numpyda bulunan np.heaviside() fonksiyonu kullanılır.
## np.heaviside(n-10) = u[n-10]
x4 = np.cos(np.pi * n / 5) * (np.heaviside(n, 1) - np.heaviside(n - 20, 1))

fig = plt.figure()
plt.stem(n, x4)
plt.title('x4[n]')
plt.xlabel('n')
plt.grid()
plt.show()

## dirac[n] = 1 için dirac adında fonksiyon oluşturuldu
def dirac(n):
    return np.where(n == 0, 1, 0)

def h2(n):
    return (dirac(n-3) - dirac(n-4))

n = np.arange(0, 25)
h = h2(n)

# konvolüsyon işlemi
y = np.convolve(x4, h)

# çıkış işareti y[n]
ny = np.arange(-5, len(y)-5)

fig = plt.figure()
plt.stem(ny, y)
plt.title('y4[n]')
plt.xlabel('n')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# giriş işareti x5[n]
n = np.arange(-2, 8)

## Hocaya sordum eğer herhangi bir fonksiyonun kullanılması yasaklanmadıysa kullanabilirsin dedi
## u[n] işareti için numpyda bulunan np.heaviside() fonksiyonu kullanılır.
## np.heaviside(n-10) = u[n-10]
x4 = (np.heaviside(n-1, 1) - np.heaviside(n-5, 1)) * (-0.3)**n

fig = plt.figure()
plt.stem(n, x4)
plt.title('x4[n]')
plt.xlabel('n')
plt.grid()
plt.show()

## dirac[n] = 1 için dirac adında fonksiyon oluşturuldu
def dirac(n):
    return np.where(n == 0, 1, 0)

def h2(n):
    return (dirac(n-3) - dirac(n-4))

n = np.arange(-3, 10)
h = h2(n)

# konvolüsyon işlemi
y = np.convolve(x4, h)

# çıkış işareti y[n]
ny = np.arange(-5, len(y)-5)

fig = plt.figure()
plt.stem(ny, y)
plt.title('y4[n]')
plt.xlabel('n')
plt.grid(True)
plt.show()
```
