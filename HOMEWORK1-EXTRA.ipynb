x[n] = 2*sin(2*pi*n/5 - pi/5) işaretinin periyodunu bulmak için aşağıdaki formül kullanılır.

x[n] = sin(w0 * n / theta) olsun. Periyot T = 2*pi/ w0'dır.

T = 2 *pi / 2*pi / 5 = 5'tir.

```python
###2 tam periyot olacak şekilde çizdirmek için np.arange(0,10) alınmıştır.
n = np.arange(0,10)

###X[n] sinyali
x_n = 2*np.sin(2*np.pi*n/5 - np.pi/5)

###Grafik çizimi
fig1=plt.figure()
ax1=fig1.add_subplot()
ax1.stem(n,x_n)
ax1.set_xlabel('n')
ax1.title.set_text('x[n]')
```

### SORU-3

## I)

X1[n] = e^(j*pi/5), n=0, n=1,n=2,...,n=15

e^(j*pi/5) = cos(pi/5) + j*sin(pi/5)'tir. İşaretin periyodunu bulabilmek için soru-2'de olduğu gibi işaretin sin yada cos n'e bağlı olmalıdır fakat işarete baktığıımızda n değerinden bağımsızdır. X1[n] değeri sabittir. w0 değeri bulunmamaktadır. Periyodu yoktur.

```python
## n değeri 0 ve 15 olmaktadır np.arange aralığı 0 ile 16 alınmıştır
n1 = np.arange(0,16)

### NOT: index uyuşmamaktadır. 0-15 arasında x1_n fonksiyonunu
###  for döngüsüne sokup n ile x1_n denkleminin indexini eşitledim

### x1 fonksiyonu tanımlanıyor
def x1_n(n):
    return np.exp(1j*np.pi / 5)

### x3 fonksiyonunun değerleri numpy array'inde tutuluyor
x1_n_real = np.array([np.real(x1_n(n1)) for n in range(16)])
x1_n_imag = np.array([np.imag(x1_n(n1)) for n in range(16)])
x1_n_amplitude = np.array([np.abs(x1_n(n)) for n in n1])
x1_n_phase = np.array([np.angle(x1_n(n)) for n in n1])
### print'in yorum satırı kaldırıldığında 0 ile 15 arasında tüm değerler 0.80901699+0.58778525j bulunmaktadır
###print(x1_n_real)
###print(x1_n_imag)

fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

ax1.stem(n1,x1_n_real)
ax1.set_xlabel('n')
ax1.title.set_text('x1[n] real')

ax2.stem(n1,x1_n_imag)
ax2.set_xlabel('n')
ax2.title.set_text('x1[n] imag')

ax3.stem(n1,x1_n_amplitude)
ax3.set_xlabel('n')
ax3.title.set_text('x1[n] amplitude')

ax4.stem(n1,x1_n_phase)
ax4.set_xlabel('n')
ax4.title.set_text('x1[n] phase')
```

## II)

x3[n] = e^(j * pi * n/5), n=0, n=1,n=2,...,n=15

e^(j*pi*n/5) = cos(pi * n/5) + j * sin(pi * n /5) 

X3[n + N] = X3[n] koşulunu sağlamaktadır. T = 2 * pi/(pi/5) = 10'dur.

```python
## n değeri 0 ve 15 olmaktadır np.arange aralığı 0 ile 16 alınmıştır
n3 = np.arange(0,16)
### NOT: index uyuşmamaktadır. 0-15 arasında x3_n fonksiyonunu
#######  for döngüsüne sokup n ile x3_n denkleminin indexini eşitledim

### x3 fonksiyonu tanımlanıyor
def x3_n(n):
    return np.exp(1j*np.pi*n / 5)

### x3 fonksiyonunun değerleri numpy array'inde tutuluyor
x3_n_real = np.array([np.real(x3_n(n)) for n in n3])
x3_n_imag = np.array([np.imag(x3_n(n)) for n in n3])
x3_n_amplitude = np.array([np.abs(x3_n(n)) for n in n3])
x3_n_phase = np.array([np.angle(x3_n(n)) for n in n3])

fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

ax1.stem(n3,x3_n_real)
ax1.set_xlabel('n')
ax1.title.set_text('x3[n] real')

ax2.stem(n3,x3_n_imag)
ax2.set_xlabel('n')
ax2.title.set_text('x3[n] imag')

ax3.stem(n3,x3_n_amplitude)
ax3.set_xlabel('n')
ax3.title.set_text('x3[n] amplitude')

ax4.stem(n3,x3_n_phase)
ax4.set_xlabel('n')
ax4.title.set_text('x3[n] phase')
```
