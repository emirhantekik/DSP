```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter("ignore")
%matplotlib notebook

##--------------------KODLAR buraya eklenecek------------------------

w_cont= np.arange(0, 2*np.pi, np.pi/1000)
X_abs=np.abs(2*np.cos(w_cont/2))
X_phase=np.array([-w/2 if w<np.pi else -w/2+np.pi for w in w_cont])
fig = plt.figure()
plt.subplot(121)
plt.plot(w_cont/np.pi,X_abs)
plt.title('$|X(e^{j\omega})|$')
plt.xlabel('Normalize frekans (x$\pi$)')
plt.subplot(122)
plt.plot(w_cont/np.pi,X_phase/np.pi)
plt.title('$ \measuredangle X(e^{j\omega})$')
plt.xlabel('Normalize frekans (x$\pi$)')
```

```python
def dirac(n):
     if n==0:
        return 1
     else:
        return 0

N=8
X_k=np.zeros([N],dtype=complex)
for k in range(N):
     for n in range(N):
        X_k[k]=X_k[k]+(dirac(n)+dirac(n-1))*np.exp(-1j*(2*np.pi/N)*k*n)
X_k_abs=np.abs(X_k)
X_k_phase=np.angle(X_k)
dif = plt.figure()
plt.figure()
plt.subplot(121)
plt.plot(w_cont/np.pi,X_abs,label='$|X(e^{j\omega})|$')
w_discrete=np.arange(0,N)*(2*np.pi/N)
plt.stem(w_discrete/np.pi,X_k_abs,'r-',label='$|X[k]|$',markerfmt='ro')
plt.xlabel('$\omega$ (x$\pi$)')
plt.legend()
plt.subplot(122)
plt.plot(w_cont/np.pi,X_phase,label='$ \measuredangle X(e^{j\omega})$')
plt.stem(w_discrete/np.pi,X_k_phase,'r-',label='$ \measuredangleX[k]$',markerfmt='ro')
plt.xlabel('$\omega$ (x$\pi$)')
plt.legend()
```

## Ön hazırlık ödevleri (ÖDEV-1 başlığı altındaki soruların çözümleri aşağıda verilecek) 

### SORU-1

h[n] = dirac[n] - dirac[n-1]

H(e^jw) = $\sum_{n=-sonsuz}^{sonsuz} h[n]e^(-jw) $
        
H(e^jw) = $\sum_{n=-sonsuz}^{sonsuz} (dirac[n]-dirac[n-1]) e^ (-jw) $

H(e^jw) = $\sum_{n=-sonsuz}^{sonsuz} dirac[n]e^(-jw) $ - $\sum_{n=-sonsuz}^{sonsuz}dirac[n-1]e^ (-jw) $

H(e^jw) = 1-e^(-jw) bulunur. Bulunan sonuç ile genlik ve faz hesaplanmayacağından dolayı farklı bir ifade bulmalıyız.H(e^jw) = 1-e^(-jw) = 2j*e^(-jw/2)((e^(jw/2)-e^(-jw/2))/2j) haline dönüştürülür ve ((e^(jw/2)-e^(-jw/2))/2j) = sin(w/2)dir. Bulunan ifade düzenlendiğinde :

H(e^jw) = 2j*e^(-jw/2)sin(w/2)

H(e^jw)'nın fazı ve genliği hesaplanırken geçen dönem alınan analog haberleşme dersindeki j = e^(j*pi/2) ifadesi yerine konur ve

|H(e^jw)| = |2sin(w/2)|

$ \measuredangle |H(e^jw)|$ = pi/2-w/2    ,0<w<pi 

            3*pi/2-w/2  ,pi<w<2*pi
