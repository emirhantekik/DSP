### SORU-2

X[n] = 2*sin(2*pi*n/5 - pi /5) + sin(3*pi*n/5 - 2*pi/5) denkleminin periyodunu bulmak için 2 sin işaretinin ayrı ayrı periyodu alınır.

x[n] = sin(w0 * n / theta) olsun. Periyot T = 2*pi/ w0'dır.

2*sin(2*pi*n/5) için T = 2*pi/(2*pi/5) => T1 = 5'dir.

sin(3*pi*n/5 - 2*pi/5) için T = 2*pi/(3*pi/5) => T2 = 10/3'dür.

X[n] periyodunu bulmak için T=EKOK(T1,T2) = 10'dur.

```python
###2 tam periyot olacak şekilde çizdirmek için np.arange(0,20) alınmıştır.
n = np.arange(0,20)

###X[n] sinyali
x_n = 2 * np.sin(2*np.pi*n/5 - np.pi/5) + np.sin(3*np.pi*n/5 -2*np.pi/5)

###Grafik çizimi

fig1=plt.figure()
ax1=fig1.add_subplot()
ax1.stem(n,x_n)
ax1.set_xlabel('n')
ax1.title.set_text('x[n]')
```

### SORU-3

X2[n] = e^(j*pi/5), n=0, n=1,n=2,...,n=15

e^(j*pi/5) = cos(pi/5) + j*sin(pi/5)'tir. sistemin periyodunu bulabilmek için soru-2'de olduğu gibi işaretin sin yada cos n'e bağlı olmalıdır fakat işarete baktığıımızda n değerinden bağımsızdır. X2[n] değeri sabittir. w0 değeri bulunmamaktadır. Periyodu yoktur.

```python
## n değeri 0 ve 15 olmaktadır np.arange aralığı 0 ile 16 alınmıştır
n2 = np.arange(0,16)

### NOT: index uyuşmamaktadır. 0-15 arasında x2_n fonksiyonunu
#######  for döngüsüne sokup n ile x2_n denkleminin indexini eşitledim

### x2 fonksiyonu tanımlanıyor
def x2_n(n):
    return np.exp(1j*np.pi / 5)

### x2 fonksiyonunun değerleri numpy array'inde tutuluyor
x2_n_real = np.array([np.real(x2_n(n)) for n in n2])
x2_n_imag = np.array([np.imag(x2_n(n)) for n in n2])
x2_n_amplitude = np.array([np.abs(x2_n(n)) for n in n2])
x2_n_phase = np.array([np.angle(x2_n(n)) for n in n2])

### print'in yorum satırı kaldırıldığında 0 ile 15 arasında tüm değerler 0.80901699+0.58778525j bulunmaktadır
###print(x2_n_real)
###print(x2_n_imag)

fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

ax1.stem(n2,x2_n_real)
ax1.set_xlabel('n')
ax1.title.set_text('x2[n] real')

ax2.stem(n2,x2_n_imag)
ax2.set_xlabel('n')
ax2.title.set_text('x2[n] imag')

ax3.stem(n2,x2_n_amplitude)
ax3.set_xlabel('n')
ax3.title.set_text('x2[n] amplitude')

ax4.stem(n2,x2_n_phase)
ax4.set_xlabel('n')
ax4.title.set_text('x2[n] phase')
```
