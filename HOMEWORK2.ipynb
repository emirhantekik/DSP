## El çözümü

Bu fark denklemi için dürtü cevabı, sistem başlangıçta sıfırda iken birim bir dürtü uygulandığında sistem çıkışındaki cevaptır. Yani, x[n]=dirac[n] olacak şekilde fark denklemi aşağıdaki gibi yazılabilir:

y[n]-1.2y[n-1]=dirac[n]-dirac[n-4]

Başlangıç sıfırda olduğundan, y[-1]=y[-2]=y[-3] = 0'dır.

#### n = 0 için:
    y[0]-1.2y[-1]=dirac[0]-dirac[-4] ==> y[0] = 1
#### n = 1 için:
    y[1]-1.2y[0]=dirac[1]-dirac[-3] ==> y[1] = 1.2
#### n= 2 için:
    y[2]-1.2y[1] = dirac[2]-dirac[-2] ==> y[2] = 1.44

```python
import matplotlib.pyplot as plt
import numpy as np

# giriş işareti x1[n]
n = np.arange(-5, 26)
x1 = np.cos(np.pi * n / 5) * (np.heaviside(n - 10, 1) - np.heaviside(n - 20, 1))

fig = plt.figure()
plt.stem(n, x1)
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.grid()
plt.show()

# sistem fonksiyonu h1[n]
def dirac(n):
    return np.where(n == 0, 1, 0)

def h1(n):
    return 1/3*(dirac(n+1) + dirac(n) + dirac(n-1))

n = np.arange(-5, 26)
h = h1(n)

fig = plt.figure(figsize=(5,3))
plt.stem(n, h)
plt.title('h1[n]')
plt.xlabel('n')
plt.ylabel('h1[n]')
plt.grid()
plt.show()

# konvolüsyon işlemi
y = np.convolve(x1, h)

# çıkış işareti y[n]
ny = np.arange(-5, len(y)-5)

fig = plt.figure()
plt.stem(ny, y)
plt.title('y[n]')
plt.xlabel('n')
plt.ylabel('value')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# giriş işareti x1[n]
n = np.arange(-5, 26)
x1 = (np.heaviside(n - 3, 1) - np.heaviside(n - 5, 1)) * (-0.3) ** n

fig = plt.figure(figsize=(8, 4))
plt.stem(n, x1)
plt.xlim([-5, 25])
plt.ylim([-0.03, 0.02])
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.title('x1[n]')
plt.grid()
plt.show()

# sistem fonksiyonu h1[n]
def dirac(n):
    return np.where(n == 0, 1, 0)

def h1(n):
    return 1/3*(dirac(n+1) + dirac(n) + dirac(n-1))

n = np.arange(-5, 26)
h = h1(n)

fig = plt.figure(figsize=(8, 4))
plt.stem(n, h)
plt.xlim([-5, 5])
plt.ylim([-0.1, 0.4])
plt.title('h1[n]')
plt.xlabel('n')
plt.ylabel('h1[n]')
plt.grid()
plt.show()

# konvolüsyon işlemi
y = np.convolve(x1, h)

# çıkış işareti y[n]
ny = np.arange(-5, len(y) - 5)

fig = plt.figure(figsize=(8, 4))
plt.stem(ny, y)
plt.xlim([0, 20])
plt.ylim([-0.01, 0.01])
plt.title('y[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)
plt.show()
```
